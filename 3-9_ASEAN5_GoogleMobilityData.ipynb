```python
#csv files: 2022 Google Mobility Data
ID_2022 = pd.read_csv('2022_ID_Region_Mobility_Report.csv')
MY_2022 = pd.read_csv('2022_MY_Region_Mobility_Report.csv')
PH_2022 = pd.read_csv('2022_PH_Region_Mobility_Report.csv')
SG_2022 = pd.read_csv('2022_SG_Region_Mobility_Report.csv')
TH_2022 = pd.read_csv('2022_TH_Region_Mobility_Report.csv')

#Display dataset shape
print(ID_2022.shape)
print(MY_2022.shape)
print(PH_2022.shape)
print(SG_2022.shape)
print(TH_2022.shape)

#View whole dataframe
#print(usdphp_1, psei_1)

#print(psei_1.to_string())
```

```python
ID_2022.rename(columns={'retail_and_recreation_percent_change_from_baseline': 'ID_retail_recreation', \
                          'grocery_and_pharmacy_percent_change_from_baseline':'ID_grocery_pharmacy', \
                                'parks_percent_change_from_baseline':'ID_parks',\
                         'transit_stations_percent_change_from_baseline':'ID_transit_stations',\
                         'workplaces_percent_change_from_baseline':'ID_workplaces',\
                         'residential_percent_change_from_baseline':'ID_residential'},inplace=True)

MY_2022.rename(columns={'retail_and_recreation_percent_change_from_baseline': 'MY_retail_recreation', \
                          'grocery_and_pharmacy_percent_change_from_baseline':'MY_grocery_pharmacy', \
                                'parks_percent_change_from_baseline':'MY_parks',\
                         'transit_stations_percent_change_from_baseline':'MY_transit_stations',\
                         'workplaces_percent_change_from_baseline':'MY_workplaces',\
                         'residential_percent_change_from_baseline':'MY_residential'},inplace=True)

PH_2022.rename(columns={'retail_and_recreation_percent_change_from_baseline': 'PH_retail_recreation', \
                          'grocery_and_pharmacy_percent_change_from_baseline':'PH_grocery_pharmacy', \
                                'parks_percent_change_from_baseline':'PH_parks',\
                         'transit_stations_percent_change_from_baseline':'PH_transit_stations',\
                         'workplaces_percent_change_from_baseline':'PH_workplaces',\
                         'residential_percent_change_from_baseline':'PH_residential'},inplace=True)

SG_2022.rename(columns={'retail_and_recreation_percent_change_from_baseline': 'SG_retail_recreation', \
                          'grocery_and_pharmacy_percent_change_from_baseline':'SG_grocery_pharmacy', \
                                'parks_percent_change_from_baseline':'SG_parks',\
                         'transit_stations_percent_change_from_baseline':'SG_transit_stations',\
                         'workplaces_percent_change_from_baseline':'SG_workplaces',\
                         'residential_percent_change_from_baseline':'SG_residential'},inplace=True)

TH_2022.rename(columns={'retail_and_recreation_percent_change_from_baseline': 'TH_retail_recreation', \
                          'grocery_and_pharmacy_percent_change_from_baseline':'TH_grocery_pharmacy', \
                                'parks_percent_change_from_baseline':'TH_parks',\
                         'transit_stations_percent_change_from_baseline':'TH_transit_stations',\
                         'workplaces_percent_change_from_baseline':'TH_workplaces',\
                         'residential_percent_change_from_baseline':'TH_residential'},inplace=True)
```

```python
TH_2022_b = TH_2022 

TH_2022_b = TH_2022.dropna(axis=1)

columns_to_drop = ['country_region_code','country_region','place_id']
TH_2022_b = TH_2022_b.drop(columns_to_drop, axis=1, inplace=False)

TH_2022_b
```

```python
SG_2022_b = SG_2022 

SG_2022_b = SG_2022.dropna(axis=1)

columns_to_drop = ['country_region_code','country_region','place_id']
SG_2022_b = SG_2022_b.drop(columns_to_drop, axis=1, inplace=False)

SG_2022_b
```

```python
PH_2022_b = PH_2022 

PH_2022_b = PH_2022_b.loc[(PH_2022_b.sub_region_1.isnull()) & (PH_2022_b.metro_area.isnull())]
PH_2022_b = PH_2022_b.dropna(axis=1)

columns_to_drop = ['country_region_code','country_region','place_id']
PH_2022_b = PH_2022_b.drop(columns_to_drop, axis=1, inplace=False)

#PH_2022_b['sub_region_1'].value_counts().dtype

PH_2022_b
```

```python
MY_2022_b = MY_2022 

MY_2022_b = MY_2022_b.loc[(MY_2022_b.sub_region_1.isnull()) & (MY_2022_b.iso_3166_2_code.isnull())]
MY_2022_b = MY_2022_b.dropna(axis=1)

columns_to_drop = ['country_region_code','country_region','place_id']
MY_2022_b = MY_2022_b.drop(columns_to_drop, axis=1, inplace=False)

MY_2022_b
```

```python
ID_2022_b = ID_2022 

ID_2022_b = ID_2022_b.loc[(ID_2022_b.sub_region_1.isnull()) & (ID_2022_b.iso_3166_2_code.isnull())]
ID_2022_b = ID_2022_b.dropna(axis=1)

columns_to_drop = ['country_region_code','country_region','place_id']
ID_2022_b = ID_2022_b.drop(columns_to_drop, axis=1, inplace=False)

ID_2022_b
```

```python
ID_2022_b = ID_2022_b.set_index('date')
MY_2022_b = MY_2022_b.set_index('date')
PH_2022_b = PH_2022_b.set_index('date')
SG_2022_b = SG_2022_b.set_index('date')
TH_2022_b = TH_2022_b.set_index('date')
```

```python
combined_data = ID_2022_b.join(MY_2022_b.join(PH_2022_b.join(SG_2022_b.join(TH_2022_b))))

combined_data
```
